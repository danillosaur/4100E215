# **IF**

```python
a = 17
b = 60
if a < b:
  print ("a is less than b")
```

# **ELIF**

```python
a = 17
b = 60
if b < a:
  print("b is greater than a")
elif a < b:
  print("a is less than b")
```

# **ELSE**

```python
a = 15
b = 15
if b < a:
  print("b is greater than a")
elif a < b:
  print("a is less than b")
else:
  print ("a and b are equal")
```

# **LEAP YEAR**

https://en.wikipedia.org/wiki/Leap_year

```python
year = eval(input("Please input the year"))

if ((year%400==0) or (year%4==0 and year%100!=0)):
  print("{0} is a leapyear".format(year))
else:
  print("{0} is not a leapyear".format(year))
```

# **range**

# **Break Statement**

```python
characters = ["Harry", "Hermione", "Ron"]
for x in characters:
  print(x)
  if x == "Hermione":
    break
```

# **Continue Statement**

```python
character = ["Harry", "Hermione", "Ron", "Albus", "Snape"]
for x in characters:
  if x == "Ron":
    continue
  print(x)
```

```python
for x in range(17):
  if x%3 == 0:
    continue
  print (x)
```

# **While Loops**

```python
b = 3
while b < 10:
  print(b)
  b += 2
```

```python
b = 3
while b < 10:
  b += 2
  if b == 6:
    continue
  print (b)
```

```python
b = 30
while b > 10:
  print(b)
  if b == 9:
    break
  b -= 2
```
